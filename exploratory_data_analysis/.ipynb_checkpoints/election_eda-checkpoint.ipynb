# Pre Processing

```python
def replace_references(text):
  us_references = [
      "US",
      "USA",
      "United States",
  ]

  harris_references = [
      "kamala harris",
      "vice president kamala harris",
      "vice president harris",
      "kamala"
  ]

  trump_references = [
      "former president donald trump",
      "former president trump",
      "donald trump",
      "donald",
  ]

  for ref in us_references:
    text = text.replace(ref, "unitedstates")
  text = text.lower()

  # Replace references one by one
  for ref in trump_references:
      text = text.replace(ref, "trump")

  for ref in harris_references:
      text = text.replace(ref, "harris")

  # Strip any leading or trailing whitespace
  cleaned_text = text.strip()

  return cleaned_text
```

```python
def preprocess_text(text):
  #clean up text
  clean_text = text.replace("“", '\"')
  clean_text = clean_text.replace("”", '\"')
  clean_text = re.sub(r'‘(.*?)’', r'beginquote \1 endquote ', clean_text)
  clean_text = clean_text.replace("’s", "\'s")
  clean_text = re.sub(r"(\w)'s", r"\1", clean_text)
  clean_text = re.sub(r'[^\w\s]', '', clean_text)

  clean_text = clean_text.replace("\\n", "\n")
  clean_text = re.sub(r'"(.*?)"', r'beginquote \1 endquote ', clean_text)
  clean_text = re.sub(r'[^\w\s]', '', clean_text)

  clean_text = replace_references(clean_text)
  #print(clean_text)

  #get tokens
  """
  nlp = spacy.load("en_core_web_sm")
  doc = nlp(text)
  cleaned_sentences = []

  cleaned_string = ""
  is_first = True
  for sentence in doc.sents:
      # Remove punctuation from each sentence
      cleaned_tokens = [token.text for token in sentence if not token.is_punct and token.text != "\n"]
      cleaned_tokens = ["not" if token=="n’t" else token for token in cleaned_tokens]
      # Join the cleaned tokens back into a string
      #print(cleaned_tokens)
      cleaned_sentence = ' '.join(cleaned_tokens)
      cleaned_sentences.append(cleaned_sentence)
      if is_first:
        cleaned_string = cleaned_sentence
        is_first = False
      else:
        cleaned_string = cleaned_string + "\n" + cleaned_sentence
  """

  #print(cleaned_string)
  #print(cleaned_string.count("\n"))

  return clean_text

#sample = df.loc[0]["Main Text"]
#raw_text = preprocess_text(sample)
#print(raw_text)
```

```python
def remove_stopwords(text):
  wnl  = WordNetLemmatizer()

  stop_words = set(stopwords.words('english'))
  filtered_text = ""
  for line in text.splitlines():
    filtered_line = ""
    for word in line.split():
      if word not in stop_words:
        filtered_line += wnl.lemmatize(word) + " "
    filtered_text += filtered_line.strip() + "\n"

  return filtered_text

#x = df_cleaned["Main Text"].iloc[0]
#y = remove_stopwords(x)
#print(y)
```

```python
columns_to_process = ["Title", "Description", "Main Text"]
df_cleaned  = pd.DataFrame(columns=columns_to_process)
df_cleaned [columns_to_process] = df[columns_to_process].copy()
df_cleaned = df_cleaned.dropna()
df_cleaned.reset_index(drop=True, inplace=True)
df_cleaned[columns_to_process] = df_cleaned[columns_to_process].apply(lambda x : x.apply(preprocess_text))
no_stopword_columns = ["new_title", "new_description", "new_main_text"]
df_cleaned[no_stopword_columns] = df_cleaned[columns_to_process].apply(lambda x : x.apply(remove_stopwords))
df_cleaned = df_cleaned.rename(columns={
    "Title" : "original_title",
    "Description" : "original_description",
    "Main Text" : "original_main_text"})

df_cleaned["outlet"] = df["ID"].str.split("-").str[0]

display(df_cleaned)
```

# Data Visualization

## Fox News

```python
def plot_wordcloud(text, heading):
  text_cleaned = text.replace("beginquote", "")
  text_cleaned = text_cleaned.replace("endquote", "")
  text_cleaned = text_cleaned.replace("say", "")
  text_cleaned = text_cleaned.replace("said", "")

  words = text_cleaned.split()

  # Calculate word frequencies
  word_freq = Counter(words)

  # Display the 10 most common words
  #print(word_freq.most_common(10))

  # Generate a word cloud
  wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

  # Display the word cloud
  plt.figure(figsize=(10, 5))
  plt.imshow(wordcloud, interpolation='bilinear')
  plt.axis('off')
  plt.title(heading)
  plt.show()
```

```python
def plot_top_words(text, title):
    # Preprocess the text (remove punctuation and lowercase)
    text_cleaned = re.sub(f"[{string.punctuation}]", "", text.lower())
    text_cleaned = text_cleaned.replace("beginquote", "")
    text_cleaned = text_cleaned.replace("endquote", "")
    text_cleaned = text_cleaned.replace("say", "")
    text_cleaned = text_cleaned.replace("said", "")
    text_cleaned = text_cleaned.replace("newsmax", "")
    text_cleaned = text_cleaned.replace("would", "")



    # Split the text into words
    words = text_cleaned.split()

    # Get word frequencies
    word_counts = Counter(words)

    # Get the 10 most common words
    most_common_words = word_counts.most_common(10)

    # Separate the words and their frequencies
    words, frequencies = zip(*most_common_words)

    # Plotting the bar graph
    plt.figure(figsize=(10, 6))
    plt.bar(words, frequencies, color='lightcoral')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Top 10 Most Frequent Words - ' + title)
    plt.xticks(rotation=45)
    plt.show()
```

```python
fox_news_title_text = df_cleaned[df_cleaned["outlet"] == "fox"]["new_title"].str.cat(sep=" ")
fox_news_main_text = df_cleaned[df_cleaned["outlet"] == "fox"]["new_main_text"].str.cat(sep=" ")
```
